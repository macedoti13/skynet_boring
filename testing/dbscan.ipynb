```python
import numpy as np
from clustering.dbscan import DBSCAN
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate dataset
X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=0.60)

# Plot dataset
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Synthetic 2D Dataset with 3 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
model = DBSCAN(max_distance=5, min_points=3)
model.fit(X)
```

```python
def visualize_dbscan(model, X):
    labels = model.predict(X)
    
    # Plotting the points with different colors for different clusters
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k', s=40)
    
    # Marking the noise points (labeled as -1) with a different marker and size
    noise_mask = (labels == -1)
    plt.scatter(X[noise_mask, 0], X[noise_mask, 1], c='red', marker='x', s=50)

    plt.title("DBSCAN Clustering")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.colorbar()
    plt.show()
```

```python
visualize_dbscan(model, X)
```
