```python
import numpy as np
from clustering.kmeans import KMeans
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate dataset
X, y = make_blobs(n_samples=300, centers=3, random_state=42, cluster_std=0.60)

# Plot dataset
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Synthetic 2D Dataset with 3 Clusters")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
model = KMeans(3)
model.fit(X)
```

```python
def visualize_kmeans(model, X):
    labels = model.predict(X)

    for i in range(model.k):
        plt.scatter(X[labels == i][:, 0], X[labels == i][:, 1], label=f'Cluster {i+1}')

    # Plot the centroids
    plt.scatter(model.centroids[:, 0], model.centroids[:, 1], c='black', marker='x', s=100, label='Centroids')
    
    plt.legend()
    plt.title('KMeans Clustering Results')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()
```

```python
visualize_kmeans(model, X)
```
